```python
# Dependencies
import requests
import json
from pprint import pprint
from config import api_key
```

```python
# url = "https://DEMO_KEY@api.nal.usda.gov/fdc/v1/search"
url = "https://api.nal.usda.gov/fdc/v1/search"
```

```python
target_url = "https://api.nal.usda.gov/fdc/v1/search"
params = {'generalSearchInput':'almond'}
headers = {"x-api-key":api_key, "Content-Type":"application/json"}
response = requests.post(target_url, headers=headers, data=json.dumps(params)).json()
pprint(response)
```
