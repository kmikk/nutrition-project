```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Get key
from config import api_key
```

```python
# Import list of foods

food_file = "resources/foods.csv"

foods = pd.read_csv(food_file, dtype ='str')

foods.head(20)
```

```python
# Set up additional columns to hold information
foods['NDBno'] = ""
foods['Name'] = ""
foods.head()
```

```python
# Find the NBDno for each food

# Set up Base URL


# Set up parameters
params = {
    "api_key": api_key,
    "lt":"g",
    "ds":'Standard Reference'
}

# food_list = foods['food']

#Loop through data
for index, row in foods.iterrows():
    
    base_url = "https://api.nal.usda.gov/ndb/search"
    params['q'] = row['food']
    params['fg'] = row['group id']
    
    # Make request
    response = requests.get(base_url, params).json()
    
#     print(json.dumps(response, indent=4))
#     break
    
    try:
        foods.loc[index, 'NDBno'] = response["list"]["item"][0]["ndbno"]
        foods.loc[index, 'Name'] = response["list"]["item"][0]["name"]

    except (KeyError):
        print(f"{row['food']} not found, moving on.")
        print(response)
```
